```python
#对分类变量grade&subGrade进行处理
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
cols = ['grade','subGrade']
for j in cols:
    Traindata[j] = lb.fit_transform(Traindata[j])
Traindata[cols].head()
#这个循环对列表 cols 中的每个元素（即列名）进行操作。
#对于每个列，fit_transform 方法将该列中的类别变量转换为独一无二的整数。这些转换后的整数随后被赋值回原始数据框 Traindata 中。
```

```python
#随机森林填补缺失的值，引入随机森林
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
empLenNotNull = Traindata.employmentLength.notnull()
columns = ['loanAmnt','grade','interestRate','annualIncome','homeOwnership','term','regionCode']
```

```python
DTC.fit(Traindata.loc[empLenNotNull,columns],Traindata.employmentLength[empLenNotNull])
print(DTC.score(Traindata.loc[empLenNotNull,columns],Traindata.employmentLength[empLenNotNull]))
```

```python
#填充
for data in [Traindata]:
    empLen_pred = DTC.predict(data.loc[:,columns])
    empLenIsNull = data.employmentLength.isnull()
    data.employmentLength[empLenIsNull] = empLen_pred[empLenIsNull]
```

```python
#转化为整数
Traindata['employmentLength']=Traindata['employmentLength'].astype(int)
```

```python
empLenNotNull_test = Testdata.employmentLength.notnull()
columns = ['loanAmnt','grade','interestRate','annualIncome','homeOwnership','term','regionCode']
```

```python
#填充test
for data_test in [Testdata]:
    empLen_pred_test = DTC.predict(data_test.loc[:,columns])
    empLenIsNull_test = data_test.employmentLength.isnull()
    data_test.employmentLength[empLenIsNull_test] = empLen_pred_test[empLenIsNull_test]
```

```python
#对issuedate和earliesCreditline进行处理——将日期转为日期格式
```

```python
Traindata.issueDate  #dtype是object，不是datetimes64，因此要改
```

```python
import datetime
```

```python
Traindata['issueDate']=pd.to_datetime(Traindata['issueDate'])
Traindata['issueDate_year']=Traindata['issueDate'].dt.year.astype('int64')
Traindata['issueDate_month']=Traindata['issueDate'].dt.month.astype('int64')

Traindata['earliesCreditLine']=pd.to_datetime(Traindata['earliesCreditLine'])
Traindata['earliesCreditLine_year']=Traindata['earliesCreditLine'].dt.year.astype('int64')
Traindata['earliesCreditLine_month']=Traindata['earliesCreditLine'].dt.month.astype('int64')

Traindata=Traindata.drop(['issueDate','earliesCreditLine'],axis = 1)
```

```python
#针对数值型变量的缺失值，用中位数填充。因为平均值受到极值的影响较大，中位数填充比较合理。
Traindata[Nu_feature] = Traindata[Nu_feature].fillna(Traindata[Nu_feature].median())
```

```python
Testdata['issueDate']=pd.to_datetime(Testdata['issueDate'])
Testdata['issueDate_year']=Testdata['issueDate'].dt.year.astype('int64')
Testdata['issueDate_month']=Testdata['issueDate'].dt.month.astype('int64')

Testdata['earliesCreditLine']=pd.to_datetime(Testdata['earliesCreditLine'])
Testdata['earliesCreditLine_year']=Testdata['earliesCreditLine'].dt.year.astype('int64')
Testdata['earliesCreditLine_month']=Testdata['earliesCreditLine'].dt.month.astype('int64')

Testdata=Testdata.drop(['issueDate','earliesCreditLine'],axis = 1)
```

```python
Nu_feature_test = list(Testdata.select_dtypes(exclude=['object']).columns) #数值型变量
```
